```python
import tensorflow as tf
import numpy as np
import data_util
import seq2seq_wrapper

# Load Data:
raw_data = data_util.load('dataset/final_dataset.csv')
data = data_util.preprocess(raw_data)

train, test, valid = data_util.split_dataset_(data)
```

```python
# Parameters:
xseq_len = train[0].shape[-1]
yseq_len = train[0].shape[-1]
xvocab_size = 26+1
yvocab_size = xvocab_size
embed_size = 300
```

```python
# Training:
model = seq2seq_wrapper.Seq2Seq(xseq_len=xseq_len,
                               yseq_len=yseq_len,
                               xvocab_size=xvocab_size,
                               yvocab_size=yvocab_size,
                               embed_size=embed_size,
                               lstm_cell_size=embed_size,
                               lstm_layers=3,
                               epoch_n = 5
                               )

sess = model.train(train, valid)

#model.test(sess, test)
```
